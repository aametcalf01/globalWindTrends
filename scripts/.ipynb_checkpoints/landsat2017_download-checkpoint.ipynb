```python
import pandas as pd
import numpy as np
import geopandas as gpd
import os
import rasterio
import rasterio.plot
from rasterio.plot import show
from rasterio.merge import merge
from rasterio.mask import mask
import matplotlib.pyplot as plt
%matplotlib inline
import glob
import fiona
from shapely.geometry import mapping
import dask.dataframe as dd
```

```python
bufferPath = "/Users/andymetcalf/Documents/research/wind/globalWindTrends/data_light/outputShapefiles/stationBuffer72_1k.shp"
landsat8shpPath = "/Users/andymetcalf/Documents/research/wind/globalWindTrends/data_light/landsat8_lookup/WRS2_descending.shp"
```

```python
#open the buffer shapefiles
buffer = gpd.read_file(bufferPath)
```

```python
#landsat 8 lookup
ls8Lookup = gpd.read_file(landsat8shpPath)
ls8Lookup.head(),type(ls8Lookup), ls8Lookup.crs
```

```python
#need to change the crs for the buffers to landsat
bufferWGS84 = buffer.to_crs(ls8Lookup.crs)
bufferWGS84.crs,ls8Lookup.crs
```

```python
#get the aws landsat 8 scenes (shoud take about a minute or so...)
# s3_scenes = pd.read_csv('http://landsat-pds.s3.amazonaws.com/c1/L8/scene_list.gz', compression='gzip')
s3_scenes = dd.read_csv('https://storage.googleapis.com/gcp-public-data-landsat/index.csv.gz', compression='gzip')


s3_scenes.head(3)
```

```python
def calc_ndvi(nir,red):
    '''Calculate NDVI from integer arrays'''
    nir = nir.astype('f4')
    red = red.astype('f4')
    ndvi = (nir - red) / (nir + red)
    return ndvi
```

```python
statNDVIPath = "/Users/andymetcalf/Documents/research/wind/globalWindTrends/data_light/stationNDVI"
tempNDVIPath = "/Users/andymetcalf/Documents/research/wind/globalWindTrends/data_light/tempNDVI"
```

```python
redBand = "_B4.TIF"
nirBand = "_B5.TIF"

buffRows = bufferWGS84.shape[0] #number of stations

counter = 0
#get the scenes from AWS
for i in range(buffRows):
    print("buffer number: ",i)
    ls8_intersection = ls8Lookup[ls8Lookup.intersects(bufferWGS84.geometry[i])]
    paths, rows = ls8_intersection["PATH"].values, ls8_intersection['ROW'].values
    
        #empty list to add the images
    bulk_list = []
    # Iterate through paths and rows
    for path, row in zip(paths, rows):

        print('Path:',path, 'Row:', row)

        # Filter the Landsat Amazon S3 table for images matching path, row, cloudcover and processing state.
        scenes = s3_scenes[(s3_scenes.path == path) & (s3_scenes.row == row) & 
                           (s3_scenes.cloudCover <= 5) & 
                           (~s3_scenes.productId.str.contains('_T2')) &
                           (~s3_scenes.productId.str.contains('_RT'))&
                           (s3_scenes.acquisitionDate.str.contains("2017"))]
        print(' Found {} images\n'.format(len(scenes)))

        # If any scenes exists, select the one that have the minimum cloudCover.
        if len(scenes):
            scene = scenes.sort_values('cloudCover').iloc[0]

            # Add the selected scene to the bulk download list.
            bulk_list.append(scene)
    
    #if there is only one scene (TIFF)
    if len(bulk_list)==1:
        #use download url and product id to download .tif
        downloadURL = bulk_list[0]["download_url"]
        redband = downloadURL[:-10]+bulk_list[0]["productId"]+redBand
        nirband = downloadURL[:-10]+bulk_list[0]["productId"]+nirBand
        
        with rasterio.open(redband)as red:
            #reproject buffer into same coordinate system as the raster
            #buffer will be used to crop the raster
            buffer2 = buffer.to_crs(red.crs)
            #use mapping function to transform coordinats into json for 
            #geopandas to use the mask function
            extent_geojson = mapping(buffer2['geometry'][i])
            #crop it!
            red_crop,red_crop_transform = mask(red,
                                          [extent_geojson],
                                          crop=True)
        with rasterio.open(nirband)as nir:
            #crop it!
            nir_crop, nir_crop_transform = mask(nir,
                                               [extent_geojson],
                                                crop=True)
            #get the metadata for the ndvi
            ndvi_meta = nir.meta.copy()
        
        #calculate NDVI   
        ndvi = calc_ndvi(nir_crop,red_crop)

                    
        #Update the new cropped affine info and the new width and height
        ndvi_meta.update({'transform':nir_crop_transform,
                          'height':ndvi.shape[1],
                          'width':ndvi.shape[2],
                          'dtype':'float32'})

#         print(ndvi_meta)
        stationName = buffer.loc[[i],["stationNum"]].values[0][0]
        outpath = statNDVIPath+"/"+stationName+"_NDVI.tif"
        #write the new data
        with rasterio.open(outpath, 'w', **ndvi_meta)as ff:
            ff.write(ndvi[0],1)

#         with rasterio.open(outpath)as src:
#             show(src,cmap = 'terrain')
    elif len(bulk_list)==0:
        continue
    else: 
        print("processing more than one image")
        for j in range(len(bulk_list)):
            #use download url and product id to download .tif
            downloadURL = bulk_list[j]["download_url"]
            redband = downloadURL[:-10]+bulk_list[j]["productId"]+redBand
            nirband = downloadURL[:-10]+bulk_list[j]["productId"]+nirBand
            with rasterio.open(redband)as red:
                #reproject buffer into same coordinate system as the raster
                #buffer will be used to crop the raster
                buffer2 = buffer.to_crs(red.crs)
                #use mapping function to transform coordinats into json for 
                #geopandas to use the mask function
                extent_geojson = mapping(buffer2['geometry'][i])
                #crop it!
                red_crop,red_crop_transform = mask(red,
                                              [extent_geojson],
                                              crop=True)
            with rasterio.open(nirband)as nir:
                #crop it!
                nir_crop, nir_crop_transform = mask(nir,
                                                   [extent_geojson],
                                                    crop=True)
                #get the metadata for the ndvi
                ndvi_meta = nir.meta.copy()

            #calculate NDVI   
            ndvi = calc_ndvi(nir_crop,red_crop)


            #Update the new cropped affine info and the new width and height
            ndvi_meta.update({'transform':nir_crop_transform,
                              'height':ndvi.shape[1],
                              'width':ndvi.shape[2],
                              'dtype':'float32'})
            outpath = tempNDVIPath+"/"+str(j)+"_NDVI.tif"
            with rasterio.open(outpath, 'w', **ndvi_meta)as ff:
                ff.write(ndvi[0],1)
            
        #now we need to merge the temp ndvi's and empty the temp folder
        NDVIs = glob.glob(tempNDVIPath+"/*.tif")
        #create an empty list for the datafiles that will be part of the mosaic
        src_files_to_mosaic = []
#         print(NDVIs)
        for f in NDVIs:
            src = rasterio.open(f)
#             print(src.meta)
#             show(src,cmap = 'terrain')
#             print(src.crs)
            src_files_to_mosaic.append(src)
        mosaic, out_transform = merge(src_files_to_mosaic)
#         show(mosaic,cmap='terrain')
        
        #copy the metadata
        out_meta = src.meta.copy()
#         print(out_meta)
        #update the metadata
        out_meta.update({"driver": "GTiff",
                         "height": mosaic.shape[1],
                         "width": mosaic.shape[2],
                         "transform": out_transform,})
        #get the station name
        stationName = buffer.loc[[i],["stationNum"]].values[0][0]
        outpath = statNDVIPath+"/"+stationName+"_NDVI.tif"
        #write the new data
        with rasterio.open(outpath, 'w', **out_meta)as ff:
            ff.write(mosaic[0],1)

#         with rasterio.open(outpath)as src:
#             show(src,cmap = 'terrain')
#             print(src.meta)
        #Delete the files in the temp folder
        fileList = os.listdir(tempNDVIPath)
        for f in fileList:
            file = tempNDVIPath+"/"+f
            os.remove(file)
         
        #close the files in src_files_to_mosaic
        for s in src_files_to_mosaic:
            s.close()
    
#     counter +=1
#     if counter>2:
#         break
```

```python
bandList = ["_B2.TIF",
            "_B3.TIF",
            "_B4.TIF",
            "_B5.TIF"]
redBand = bandList[2]
nirBand= bandList[3]

buffRows = bufferWGS84.shape[0] #number of rows
counter = 0
for i in range(buffRows):
    ls8_intersection = ls8Lookup[ls8Lookup.intersects(bufferWGS84.geometry[i])]
    paths, rows = ls8_intersection["PATH"].values, ls8_intersection['ROW'].values
    
    #empty list to add the images
    bulk_list = []
    # Iterate through paths and rows
    for path, row in zip(paths, rows):

        print('Path:',path, 'Row:', row)

        # Filter the Landsat Amazon S3 table for images matching path, row, cloudcover and processing state.
        scenes = s3_scenes[(s3_scenes.path == path) & (s3_scenes.row == row) & 
                           (s3_scenes.cloudCover <= 5) & 
                           (~s3_scenes.productId.str.contains('_T2')) &
                           (~s3_scenes.productId.str.contains('_RT'))&
                           (s3_scenes.acquisitionDate.str.contains("2017"))]
        print(' Found {} images\n'.format(len(scenes)))

        # If any scenes exists, select the one that have the minimum cloudCover.
        if len(scenes):
            scene = scenes.sort_values('cloudCover').iloc[0]

        # Add the selected scene to the bulk download list.
        bulk_list.append(scene)
    if len(bulk_list)==1:
        downloadURL = bulk_list[0]["download_url"]
        redband = downloadURL[:-10]+bulk_list[0]["productId"]+redBand
        nirband = downloadURL[:-10]+bulk_list[0]["productId"]+nirBand
        with rasterio.open(redband)as src:
            oviews = src.overviews(1)
            print(oviews, src.height, src.width)
            oview = oviews[0] #highest resolution overview
            red = src.read(1, out_shape=(1, int(src.height // oview), int(src.width // oview)))
        with rasterio.open(nirband)as src:
            oviews = src.overviews(1)
            oview = oviews[0] #highest resolution overview
            nir = src.read(1, out_shape=(1, int(src.height // oview), int(src.width // oview)))  
        ndvi = calc_ndvi(nir,red)
        stationName = bufferWGS84.loc[[i],["stationNum"]].values[0][0]
        outpath = statNDVIPath+"/"+stationName+"_NDVI.tif"
        with rasterio.open(nirband) as src:
            profile = src.profile.copy()
            
            aff = src.transform
            newaff = rasterio.Affine(aff.a * oview, aff.b, aff.c,
                                     aff.d, aff.e * oview, aff.f)
            profile.update({
                    'dtype': 'float32',
                    'height': ndvi.shape[0],
                    'width': ndvi.shape[1],
                    'transform': newaff})  
            
            with rasterio.open(outpath, 'w', **profile) as dst:
                dst.write_band(1,ndvi)
    else:
        print("There are more than one image")
        for j in range(len(bulk_list)):
            downloadURL = bulk_list[j]["download_url"]
            redband = downloadURL[:-10]+bulk_list[j]["productId"]+redBand
            nirband = downloadURL[:-10]+bulk_list[j]["productId"]+nirBand
            with rasterio.open(redband)as src:
                oviews = src.overviews(1)
                oview = oviews[0] #second highest resolution overview
                oview = 1
                red = src.read(1, out_shape=(1, int(src.height // oview), int(src.width // oview)))
            with rasterio.open(nirband)as src:
                oviews = src.overviews(1)
                oview = oviews[0] #second highest resolution overview
                oview = 1
                nir = src.read(1, out_shape=(1, int(src.height // oview), int(src.width // oview)))  
            ndvi = calc_ndvi(nir,red)
            outpath = tempNDVIPath+"/"+str(j)+"_NDVI.tif"
            with rasterio.open(nirband) as src:
                profile = src.profile.copy()
            
                aff = src.transform
                newaff = rasterio.Affine(aff.a * oview, aff.b, aff.c,
                                         aff.d, aff.e * oview, aff.f)
                profile.update({
                        'dtype': 'float32',
                        'height': ndvi.shape[0],
                        'width': ndvi.shape[1],
                        'transform': newaff}) 
                with rasterio.open(outpath, 'w', **profile) as dst:
                    dst.write_band(1,ndvi)
        #Now we need to merge the temp ndvi's and empty the temp folder
        NDVIs = glob.glob(tempNDVIPath+"/*.tif")
        #create an empty list for the datafiles that will be part of the mosaic
        src_files_to_mosaic = []
        for f in NDVIs:
            src = rasterio.open(f)
            show(src,cmap = 'terrain')
            print(src.crs)
            src_files_to_mosaic.append(src)
        print(src_files_to_mosaic)
        print(tuple(buffer.bounds.loc[i]))
        print(src_files_to_mosaic[0].bounds)
        buffer2 = buffer.to_crs(src_files_to_mosaic[0].crs)
        mosaic, out_trans = merge(src_files_to_mosaic,bounds = tuple(buffer2.bounds.loc[i]))
#         mosaic, out_trans = merge(src_files_to_mosaic,bounds = src_files_to_mosaic[0].bounds)
        show(mosaic, cmap = 'terrain')
            
    counter+=1
    if counter>2:
        break
    
    
    
```
